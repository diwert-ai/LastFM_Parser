```python
import pandas as pd
import csv
df = pd.read_csv('lastfm2.csv', quotechar='|', quoting=csv.QUOTE_ALL, delimiter=' ')
len(df)
```

```python
df['code'] = (df['album page'] - 1) * 20 + df['album num']
tags = ['rock', 'hip-hop', 'indie', 'jazz', 'reggae', 'british', 'punk', '80s', 'dance', 'electronic', 'metal',
            'acoustic', 'rnb', 'hardcore', 'country', 'blues', 'alternative', 'classical', 'rap', 'composer',
            'modern classical', 'neoclassical', 'post-punk', 'russian']
uniq_s = 0
for n, tag in enumerate(tags,1):
    uniq = len(df[df['album tag'] == tag]['code'].unique())
    uniq_s += uniq
    print(f'{uniq} {uniq/500: .2f} {uniq_s/n: .0f} {uniq_s/n/500: .2f} {tag}')
```

```python
dfs = df.sort_values(by='track scrobbles', ascending=False)
dfs.filter(items=['album tag', 'album page', 'album release date',
                      'album name', 'album artist', 'track name', 'track listeners', 'track scrobbles'])[30:60]
```

```python
df['artist_track_name'] = df['album artist'] + '_' + df['track name']
len(df['artist_track_name'].unique())
```
